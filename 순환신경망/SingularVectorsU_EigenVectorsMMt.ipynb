```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def rotation(radian):
    new_axis = np.array([[np.cos(radian), -np.sin(radian)], [np.sin(radian), np.cos(radian)]])
    new_axis = np.round(new_axis, 2)
    return new_axis

def horizontal_shear(slope):
    new_axis= np.array([[1, slope], [0, 1]])
    return new_axis

def vertical_shear(slope):
    new_axis = np.array([[1, 0], [slope, 1]])
```

```python
x = np.array([[0, 0.9],[0, 0.1]])
```

```python
fig, axs = plt.subplots(nrows=1, ncols=3, figsize=(15, 5))

plt.sca(axs[0])
# 45degree(np.pi/4 radian) rotation
rot_45 = np.dot(rotation(np.pi/4), x)
plt.plot(x[0], x[1], c='red', label='0     radian')
plt.plot(rot_45[0], rot_45[1], c='green', label='pi/4 radian')

rot_90 = np.dot(rotation(np.pi/2), x)
plt.plot(rot_90[0], rot_90[1], c='blue', label='pi/2 radian')
plt.grid()
plt.xlim([-2, 2])
plt.ylim([-2, 2])
plt.legend(fontsize=10, loc='lower left')
```
