```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.array([0.9, 0.1])
```

```python
def new_axis(r):
    return np.round(np.array([[np.cos(r), -np.sin(r)],[np.sin(r), np.cos(r)]]), 2)
```

```python
new_axis(np.pi/2),  np.dot(new_axis(np.pi/2), x)
```

```python
m_90 = np.dot(new_axis(np.pi/2), x)
print(m_90)
```

```python
m_45 = np.dot(new_axis(np.pi/4), x)
m_45
```

```python
plt.figure(figsize=(5,5))
plt.plot([0, x[0]],[0, x[1]], c='red')
plt.plot([0, m_90[0]], [0, m_90[1]], c='green')
plt.plot([0, m_45[0]], [0, m_45[1]], c='purple')
plt.xlim([-1, 1])
plt.ylim([-1, 1])
plt.grid()

plt.show()
```
