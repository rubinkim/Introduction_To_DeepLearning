```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.array([[0, 0.9],[0, 0.1]])
```

```python
def new_axis(r):
    return np.round(np.array([[np.cos(r), -np.sin(r)],[np.sin(r), np.cos(r)]]), 2)
```

```python
def horizontal_shear(m):
    return np.array([[1, 0], [m, 1]])
```

```python
new_axis(np.pi/2),  np.dot(new_axis(np.pi/2), x)
```

```python
m_90 = np.dot(new_axis(np.pi/2), x)
print(m_90)
```

```python
m_45 = np.dot(new_axis(np.pi/4), x)
m_45
```

```python
hs_2 = np.dot(horizontal_shear(2), x)
hs_2, x
```

```python
plt.figure(figsize=(7,7))
plt.plot(x[0], x[1], c='red')
plt.plot(m_45[0], m_45[1], c='purple')
plt.plot(m_90[0], m_90[1], c='orange')

plt.plot(hs_2[0], hs_2[1], c='blue')
plt.xlim([-1, 1])
plt.ylim([-1, 1])
plt.grid()
```
