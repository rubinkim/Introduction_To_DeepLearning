```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
m = np.array([[-1/4, -1],[1, 1]])
```

```python
u = np.array([[1/np.sqrt(2), 1/np.sqrt(2)], [-1/np.sqrt(2), -1/np.sqrt(2)]])
```

```python
mu = np.dot(m, u)
print(mu)
```

```python
plt.plot(m[:, 0], m[:, 1], c='red')
plt.plot(u[:, 0], u[:, 1], c='blue', alpha=0.2)
plt.plot(mu[:, 0], mu[:, 1], c='green', alpha=1)
plt.xlim([-1.5, 1.5])
plt.ylim([-1.5, 1.5])
plt.show()
```
