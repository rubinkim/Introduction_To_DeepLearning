```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.array([1/2, 1/3])
```

```python
def new_axis(r):
    return np.round(np.array([[np.cos(r), -np.sin(r)],[np.sin(r), np.cos(r)]]), 2)
```

```python
new_axis(0),  np.dot(new_axis(np.pi/2), np.array([1/2, 1/3]))
```

```python
mx = np.dot(new_axis(np.pi/2), x)
print(mx)
```

```python
plt.figure(figsize=(10,10))
plt.plot(x, c='red')
plt.plot(mx, c='green')

plt.show()
```
